```python
# Pie Charts
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]

plt.pie(amounts, labels=expenses)

# set style
#plt.style.use("ggplot")
plt.style.use("default")
# add %-age of each slice
plt.pie(amounts, labels=expenses,  autopct='%1.1f%%')

#set colors
#colors = ["purple", "green", "blue", "pink", "yellow"]
#plt.pie(amounts, labels=expenses,  autopct='%1.1f%%', colors=colors)
explodes=[0,0,0,0.2,0]
# wedgeprops
plt.pie(amounts, labels=expenses,  autopct='%1.1f%%', 
        wedgeprops={"edgecolor":"black", "linewidth": 1, "linestyle":"--", "fill":True},
        explode=explodes,
        shadow=True,
        startangle=25)
#plt.savefig("figpie.png")
        
```

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Delhi", "Tokyo"]

temperatures = [
    [22, 23, 21, 24, 25],   # New York
    [18, 19, 17, 20, 21],   # London
    [30, 32, 31, 33, 34],   # Delhi
    [25, 26, 24, 27, 28]    # Tokyo
]


fig,ax=plt.subplots(2,2)
cont=0
for i in range(2):
    for j in range(2):
        ax[i][j].plot(days,temperatures[cont])
        ax[i][j].set_title(cities[cont])
        ax[i][j].grid(True)
        cont=cont+1

fig.suptitle("temp over week indiff cities")
fig.supxlabel("days")
fig.supylabel("temp")
fig.tight_layout()
```

```python
fig,axs=plt.subplots(2,2)
axs=axs.flatten()

for i,ax in enumerate(axs):
    ax.plot(days,temperatures[i])
    ax.set_title(cities[i])
    ax.grid(True)

fig.suptitle("temp of diff cities over weeks")
fig.supxlabel("Days")
fig.supylabel("Temp")
fig.tight_layout()
```
