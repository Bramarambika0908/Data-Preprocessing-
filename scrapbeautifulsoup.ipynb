```python

from bs4 import BeautifulSoup


page_count= 1

while True:
    URL= f"https://quotes.toscrape.com/page/{page_count}/"
    res= requests.get(URL)

    soup=BeautifulSoup(res.text,"lxml")
    quotes=soup.select("div.quote")

    if not quotes:
        print("no more pages!")
        break
    with open (f"scrape_data/activedata{page_count}.html","w", encoding="utf-8") as f:
        f.write(res.text)
        print(f"download done {page_count}")
    #for q in quotes:
        #text= quotes.select_one("span.text")
        #author=quotes.select_one("")
        
    page_count = page_count + 1
    
```

```python
#Extrat 
life_quotes = []
pag_c=1
while pag_c< page_count:
    with open (f"scrape_data/activedata{pag_c}.html","r", encoding="utf-8")as f:
        html_file= f.read()

        
    
    soup1= BeautifulSoup(html_file,"lxml")
    
    all_quotes = soup1.select("div.quote")
    
    for tag in all_quotes:
        all_tags=[q.get_text() for q in tag.select(".tags .tag")]
        
        if "life" in all_tags:
            text = tag.select_one("span.text").get_text()
            author = tag.select_one("small.author").get_text()
            life_quotes.append([text, author])
    pag_c= pag_c+1
print(life_quotes)
```

```python
import pandas as pd

pdf=pd.DataFrame(life_quotes, columns=["quotes","authors"])
```

```python
pdf.to_csv("scrapweb.csv")
```
